<h2>
    Obtenemos la formula
</h2>

```python
#independiente
x = datos.altura
#dependiente
y =datos.tirosCampoExito
m , b = np.polyfit(x,y,1)
```

```python
print ('el valor de pendiente es..:  {:.2f}.'.format(m))
print ('la intercepcion es..:  {:.2f}.'.format(b))
print ('la formula es {:.2f}x + {:.2f}.'.format(m,b))
```

La Formula Es : 
\begin{equation}
y(x) = 0.06x + 0.05
\end{equation}

```python
# valor medio de x e y
y_medio = datos.tirosCampoExito.mean()
# x coordernada
x_medio = datos.altura.mean()
print ('el centro para este dataset es x = {:.2f} and y = {:.2f}.'.format(y_medio, y_medio))
```

<h2>
    Graficamos la regresion con los valores
</h2>

```python
plt.figure(figsize=(15,7))
plt.plot(x,y, 'o', markersize = 5, label = "Valores")
plt.plot(x_medio,y_medio, 'o', markersize = 30, color = "r")
plt.plot(x, m*x + b, '-', label = 'Regresion Lineal', linewidth = 4)
plt.ylabel('Tiros de Campo Exitoso', fontsize = 15)
plt.xlabel('Altura(pies)', fontsize = 15)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.legend(loc = 'upper left', fontsize = 15)
```

<h2>
    Predecimos con una funcion
</h2>

```python
#creamos una funcion que predice
def prediccion(altura):
    #convertimos de centimetros a pies
    pies = altura/0.3048
    if pies >= 3.93701:
        predi = m*pies+b
        return "su porcentaje de tiros de campo exitosos es: {0:.2f}".format(predi)
    else:
        return "tu altura tiene que ser mayor a 1.20"
```
